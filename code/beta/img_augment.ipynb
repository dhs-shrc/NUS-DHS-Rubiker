Image augmentation by making pictures ready for ML training

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import glob
import datetime
import random
from tqdm.notebook import tqdm
from PIL import Image
from PIL import ImageEnhance
np.random.seed(1) #to have reproducible results 
pd.set_option('display.max_colwidth', None)
```

```
from google.colab import drive
drive.mount('/gdrive')
paths = glob.glob('/gdrive/My Drive/Projects/Rubik/dataset/*.png', recursive = True)
print(len(paths))
```

```
orig = []
i = 0
for img in paths:
    orig.append(np.asarray(Image.open(img)))
    if i%20==0:
        print(i)
    i += 1
orig = np.array(orig)
orig.shape
```

```
contrast = np.array([np.asarray(ImageEnhance.Contrast(Image.fromarray(img, 'RGB')).enhance(np.random.choice(np.linspace(0.7, 1.3, 5)))) for img in orig])
```

```
brightness = np.array([np.asarray(ImageEnhance.Brightness(Image.fromarray(img, 'RGB')).enhance(np.random.choice(np.linspace(0.5, 1.5, 5)))) for img in contrast])
sharpness = np.array([np.asarray(ImageEnhance.Sharpness(Image.fromarray(img, 'RGB')).enhance(np.random.choice(np.linspace(0.5, 1.5, 5)))) for img in brightness])
```

```
import os
for i, np_img in enumerate(sharpness):
    np_img_bgr = cv2.cvtColor(np_img, cv2.COLOR_RGB2BGR)
    cv2.imwrite('/gdrive/My Drive/Projects/Rubik/augment/'+paths[i].split('/')[6][:-4]+'_aug1.png', np_img_bgr)
```
