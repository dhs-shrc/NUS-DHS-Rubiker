```
import cv2
import keras
import numpy as np
from PIL import Image
import tensorflow as tf
import random
from keras.models import Sequential

from keras.layers import Convolution2D

from keras.layers import MaxPooling2D

from keras.layers import Flatten

from keras.layers import Dense
```

```
model = Sequential()

model.add(Dense(units = 3, activation = 'relu'))

model.add(Dense(units = 32, activation = 'relu'))

model.add(Dense(units = 6, activation = 'sigmoid'))

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```
img = Image.open('test.png')
img_array = tf.keras.preprocessing.image.img_to_array(img)
height, width, channels = img_array.shape
data = []
for i in range(height):
    for j in range(width):
        if sum(img_array[i][j])<120:
            continue
        if i<float(height)*2/3 and j<float(width)/3:  # yellow
            data.append((img_array[i][j], [0 if x!=2 else 1 for x in range(6)]))
        elif i<float(height)/3 and j<float(width)*2/3:  # blue
            data.append((img_array[i][j], [0 if x!=5 else 1 for x in range(6)]))
        elif i<float(height)*2/3 and j<float(width)*2/3:  # green
            data.append((img_array[i][j], [0 if x!=0 else 1 for x in range(6)]))
        elif i<float(height)*2/3:  # white
            data.append((img_array[i][j], [0 if x!=1 else 1 for x in range(6)]))
        elif j<float(width)*2/3:  # orange
            data.append((img_array[i][j], [0 if x!=4 else 1 for x in range(6)]))
        else:
            data.append((img_array[i][j], [0 if x!=3 else 1 for x in range(6)]))
```

```
random.seed(1)
random.shuffle(data)
X = np.array([sample[0] for sample in data])
y = np.array([sample[1] for sample in data])
print(data[0][1],y[0])
print(X[0].shape)
print(len(data))
```

```
history=model.fit(X, y, batch_size=200, epochs=5, validation_split = 0.1, verbose=1)
```
